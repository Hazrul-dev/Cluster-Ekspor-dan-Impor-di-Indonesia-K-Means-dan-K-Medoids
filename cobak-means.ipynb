Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, davies_bouldin_score
from sklearn.metrics.pairwise import pairwise_distances
```

Memanggil Data yang Digunakan

```python
df = pd.read_csv('EksImp_Dataset_BPS.csv')
data = df[['Nilai Ekspor (USD) (2021)', 'Berat Ekspor (Kg) (2021)', 'Nilai Ekspor (USD) (2022)', 'Berat Ekspor (Kg) (2022)', 'Nilai Ekspor (USD) (2023)', 'Berat Ekspor (Kg) (2023)','Nilai Impor (USD) (2021)', 'Berat Impor (Kg) (2021)', 'Nilai Impor (USD) (2022)', 'Berat Impor (Kg) (2022)', 'Nilai Impor (USD) (2023)', 'Berat Impor (Kg) (2023)']]
data.head()
```

Pre-Processing Data Menggunakan StandardScaler

```python
scaler = StandardScaler()
data_skalered = scaler.fit_transform(data)
data_df = pd.DataFrame(data_skalered, columns=['Nilai Ekspor (USD) (2021)', 'Berat Ekspor (Kg) (2021)', 'Nilai Ekspor (USD) (2022)', 'Berat Ekspor (Kg) (2022)', 'Nilai Ekspor (USD) (2023)', 'Berat Ekspor (Kg) (2023)','Nilai Impor (USD) (2021)', 'Berat Impor (Kg) (2021)', 'Nilai Impor (USD) (2022)', 'Berat Impor (Kg) (2022)', 'Nilai Impor (USD) (2023)', 'Berat Impor (Kg) (2023)',])
data_df.head(99)
```

Menggunakan Metode Elbow yaitu WCSS untuk melihat K (Centroid) terbaik

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=0 )
    kmeans.fit(data_skalered)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number Of Cluster')
plt.ylabel('WCSS_List')
plt.show()
```

Proses K-Means dan Menghitung Jarak Menggunakan Euclidean Distance

```python
kmeans = KMeans(n_clusters=3, init='k-means++',n_init=99, max_iter=300, random_state=0)
y_predict= kmeans.fit_predict(data_df)
jarak = pairwise_distances(data_df, metric='euclidean')
print(kmeans.labels_)
```

```python
plt.scatter(data_skalered[y_predict == 0,0], data_skalered[y_predict == 0,1], s = 100, c = 'blue', label ='Cluster 0')
plt.scatter(data_skalered[y_predict == 1,0], data_skalered[y_predict == 1,1], s = 200, c = 'green', label ='Cluster 1')
plt.scatter(data_skalered[y_predict == 2,0], data_skalered[y_predict == 2,1], s = 100, c = 'red', label ='Cluster 2')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s =100, c = 'yellow', label = 'Centroid')
plt.title('Clusterization of Exports and Imports of Goods in Indonesia')
plt.xlabel('Export Value and Weight')
plt.ylabel('Import Value and Weight')
plt.legend()
plt.show()
```

```python
print("Hasil klastering:")
for i in range(len(y_predict)):
    print(f"Data ke-{i+1}: {y_predict[i]}")
```

Use of DBI for Cluster Evaluation

```python
dbi = davies_bouldin_score(data_skalered,kmeans.labels_)

print(f"Indeks Davies-Bouldin: {dbi:.2f}")
```
